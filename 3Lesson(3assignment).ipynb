# **Queue**

first-in, first-out (FIFO) 
![alt text](https://python-scripts.com/wp-content/uploads/2017/05/queues-python-1024x575.jpg)

```
#Будем работать со списком. 
class Queue:


  def __init__(self):
    self.our_queue = []

  def get_queue_len(self):
    return f'The len of this queue is: {len(self.our_queue)}' 
    

  def in_queue(self, number):
    self.our_queue.insert(0, number)
    return f'Value:{number} in'
    

  def out_queue(self):
    _out = self.our_queue.pop()
    print(f'Value:{_out} out')
    


q = Queue()

print(q.get_queue_len())
print(q.in_queue(5))
print(q.in_queue(7))
print(q.in_queue('www'))

print(q.get_queue_len())
q.out_queue()
q.out_queue()

```

## Стек. Last-In/First-Out

```

class Stack:


  def __init__(self):
    self.our_stack = []

  def get_stack_len(self):
    return f'The len of this stack is: {len(self.our_stack)}' 
    

  def in_stack(self, number):
    self.our_stack.append(number)
    return f'Value:{number} in'
    

  def out_stack(self):
    _out = self.our_stack.pop()
    print(f'Value:{_out} out')
    


q = Stack()

print(q.in_stack(3))
print(q.in_stack(10))
print(q.in_stack('fff'))
q.out_stack()
```

# **Класс комплексного числа.**
комплексное число принято записывать именно в таком порядке:  ![alt text](http://mathprofi.ru/h/kompleksnye_chisla_dlya_chainikov_clip_image008_0000.gif)

Создать класс комплексного числа и реализовать для него арифметические операции.

```
#Complex number class

import matplotlib.pyplot as plt
import numpy as np

class CompNum:
  a_list = []
  b_list = []
  set_sum_a = []
  set_sum_b = []

  def __init__(self, a, b):
    self.a_list.append(a)
    self.b_list.append(b)
    

  def set_a(self, a):
    self.a_list.append(a)
     

  def set_b(self, b):
    self.b_list.append(b)


  def get_a(self):
    return f'Our list a:{self.a_list}' 


  def get_b(self):
    return f'Our list b:{self.b_list}' 

  def sum_comnum(self, idx_first_comnum, idx_second_comnum):
    #(a+bi)+(c+di)=(a+c)+(b+d)i
    self.comnum_summ_a = self.a_list[idx_first_comnum] + self.a_list[idx_second_comnum]
    self.comnum_summ_b = self.b_list[idx_first_comnum] + self.b_list[idx_second_comnum]
    if self.comnum_summ_a == 0:
      return f'Our complex number diff:z = {self.comnum_summ_b}i'  
    if self.comnum_summ_b == 0:
      return f'Our complex number diff:z = {self.comnum_summ_a}'  
    if self.comnum_summ_b == 0 and self.comnum_summ_a == 0:   
      return f'Our complex number diff:z = 0'
    if self.comnum_summ_b < 0:
      return f'Our complex number diff:z = {self.comnum_summ_a} {self.comnum_summ_b}i'
    else:
        return f'Our complex number diff:z = {self.comnum_summ_a} + {self.comnum_summ_b}i'
    return f'Our complex number sum:z = {self.comnum_summ_a} + {self.comnum_summ_b}i'

  def diff_comnum(self, idx_first_comnum, idx_second_comnum):
  #От первой точки отнимаем вторую!    
    #(a+bi)−(c+di)=(a−c)+(b−d)i
    self.comnum_diff_a = self.a_list[idx_first_comnum] - self.a_list[idx_second_comnum]
    self.comnum_diff_b = self.b_list[idx_first_comnum] - self.b_list[idx_second_comnum]
    if self.comnum_diff_a == 0:
      return f'Our complex number diff:z = {self.comnum_diff_b}i'  
    if self.comnum_diff_b == 0:
      return f'Our complex number diff:z = {self.comnum_diff_a}'  
    if self.comnum_diff_b == 0 and self.comnum_diff_a == 0:   
      return f'Our complex number diff:z = 0'
    if self.comnum_diff_b < 0:
      return f'Our complex number diff:z = {self.comnum_diff_a} {self.comnum_diff_b}i'
    else:
        return f'Our complex number diff:z = {self.comnum_diff_a} + {self.comnum_diff_b}i'


  def mult_comnum(self, idx_first_comnum, idx_second_comnum):
    #(a+bi)(c+di)=(ac−bd)+(ad+bc)i
    self.comnum_mult_a = (self.a_list[idx_first_comnum] * self.a_list[idx_second_comnum] 
    - self.b_list[idx_first_comnum] * self.b_list[idx_second_comnum])
    self.comnum_mult_b = (self.a_list[idx_first_comnum] * self.b_list[idx_second_comnum] 
    + self.b_list[idx_first_comnum] * self.a_list[idx_second_comnum])
    if self.comnum_mult_a == 0:
      return f'Our complex number mult:z = {self.comnum_mult_b}i'  
    if self.comnum_mult_b == 0:
      return f'Our complex number mult:z = {self.comnum_mult_a}'  
    if self.comnum_mult_b == 0 and self.comnum_mult_a == 0:   
      return f'Our complex number mult:z = 0'
    if self.comnum_mult_b < 0:
      return f'Our complex number mult:z = {self.comnum_mult_a} {self.comnum_mult_b}i'
    else:
        return f'Our complex number mult:z = {self.comnum_mult_a} + {self.comnum_mult_b}i'

  def dev_comnum(self, idx_first_comnum, idx_second_comnum):
#     a+bi     ac(2*1)+bd(5*3)          bc(5*1) -  ad(2*3)     17    -1
#    ------ = ----------------    +  ----------------------i = -- + ----i 
#     c+di    c^2(1*1) + d^2(3*3)       c^2(1*1) + d^2(3*3)    10    10 

    self.comnum_dev_a = (self.a_list[idx_first_comnum] * self.a_list[idx_second_comnum]
    + self.b_list[idx_first_comnum] * self.b_list[idx_second_comnum]) / (
    self.a_list[idx_second_comnum] * self.a_list[idx_second_comnum] +
    self.b_list[idx_second_comnum] * self.b_list[idx_second_comnum])
    self.comnum_dev_b = (self.b_list[idx_first_comnum] * self.a_list[idx_second_comnum]
    - self.a_list[idx_first_comnum] * self.b_list[idx_second_comnum]) / (
    self.a_list[idx_second_comnum] * self.a_list[idx_second_comnum] +
    self.b_list[idx_second_comnum] * self.b_list[idx_second_comnum])
    if self.comnum_dev_a == 0:
      return f'Our complex number dev:z = {self.comnum_dev_b}i'  
    if self.comnum_dev_b == 0:
      return f'Our complex number dev:z = {self.comnum_dev_a}'  
    if self.comnum_dev_b == 0 and self.comnum_dev_a == 0:   
      return f'Our complex number dev:z = 0'
    if self.comnum_dev_b < 0:
      return f'Our complex number dev:z = {self.comnum_dev_a} {self.comnum_dev_b}i'
    else:
        return f'Our complex number dev:z = {self.comnum_dev_a} + {self.comnum_dev_b}i'


  def comnum_graph(self):
    #Если успею - сделать из него метод с классовым декоратором    
    x = self.a_list
    y = self.b_list
    fig, ax = plt.subplots(figsize=(6, 6))
    ax.set_title("Complex numbers")
    ax.set_xlabel('Re', fontsize=16)
    ax.set_ylabel('Im', fontsize=16)
    ax.grid()
    
    ax.plot(x, y, 'r.', label='z=a+bi')
    ax.plot(self.comnum_summ_a, self.comnum_summ_b,'b.', label='common numbers - summ')
    ax.plot(self.comnum_diff_a, self.comnum_diff_b,'g.', label='common numbers - difference')
    ax.plot(self.comnum_mult_a, self.comnum_mult_b,'y.', label='common numbers - multiplication')
    ax.plot(self.comnum_dev_a, self.comnum_dev_b,'y.', label='common numbers - devide')

    ax.legend()
    plt.show()




comp_number = CompNum(2,5)
comp_number.set_a(1)
comp_number.set_b(3)


print(comp_number.get_a())
print(comp_number.get_b())

print(comp_number.sum_comnum(0, 1))
print(comp_number.diff_comnum(0, 1))
print(comp_number.mult_comnum(0, 1))
print(comp_number.dev_comnum(0, 1))
comp_number.comnum_graph()

```
